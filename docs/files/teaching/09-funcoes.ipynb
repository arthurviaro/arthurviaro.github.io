Até aqui, nossos programas seguiam um fluxo linear: o Python executava cada instrução na ordem em que ela aparecia. Essa abordagem funciona bem para tarefas simples, mas rapidamente se torna problemática à medida que os programas crescem.

Imagine que você precisa calcular o PIB pela ótica da despesa em vinte cenários diferentes. Sem funções, você repetiria a mesma fórmula vinte vezes — e bastaria um erro em uma das cópias para comprometer toda a análise.

**Funções** resolvem esse problema: elas permitem que um bloco de código seja escrito uma única vez e reutilizado quantas vezes for necessário, com diferentes valores de entrada.

## O que são funções? <a id="funcoes"></a>

- Funções são blocos de código reutilizáveis que executam uma tarefa específica.
- Permitem organizar, reutilizar e dividir o código em partes menores e mais compreensíveis.

### Por que usar funções?

- Evitam repetição de código.
- Facilitam a manutenção e o entendimento do programa.
- Permitem reutilização e modularização.

### Sintaxe básica de uma função

- Para definir uma nova função no Python, usamos a palavra-chave `def`, seguida do nome da função e, opcionalmente, pelos parâmetros.
- Entre parênteses, definimos os **parâmetros** que a função recebe.
- Todo o bloco de código **indentado** abaixo do cabeçalho da função fará parte dela.
- Embora não seja obrigatório, as funções podem retornar algum resultado.

A estrutura básica de uma função é a seguinte:

- `def`: Palavra-chave que indica a definição de uma função.
- `nome_da_funcao`: Nome descritivo que indica o que a função faz.
- `parametros`: Valores que a função recebe para processar (opcional).
- `return`: Palavra-chave que indica o valor que a função devolve após sua execução (opcional).

```python
def nome_da_funcao(parametros):
    <instrucoes>
    ...
    return <valor> (opcional)
```

````python
def saudacao():
    print("Olá, seja bem-vindo(a)!")

saudacao()
````

## Parâmetros de uma função <a id="parametros"></a>

- Os **parâmetros** tornam as funções flexíveis, permitindo que diferentes valores sejam processados sem a necessidade de alterar o código da função.
- Isso facilita a reutilização, já que o mesmo bloco de código pode ser utilizado em várias partes do programa, com diferentes valores de entrada.

````python
def saudacao(nome):
    print(f"Olá, {nome}! Seja bem-vindo(a)!")

nome = input("Qual é o seu nome? ").strip().title()
saudacao(nome)
````

- No exemplo anterior, estamos dizendo ao interpretador que a função `saudacao()` recebe um único parâmetro: uma variável chamada `nome`.
- Portanto, quando usamos `saudacao(nome)`, o Python passa o valor da variável nome para a função `saudacao()` como `nome`.
- Podemos modificar o nosso código para adicionar um **valor padrão** à função:

````python
def saudacao(nome = 'visitante'):
    print(f"Olá, {nome}! Seja bem-vindo(a)!")

nome = input("Qual é o seu nome? ").strip().title()
saudacao(nome)

saudacao()
````

- Note que, como nenhum valor é passado na segunda vez em que a função `saudacao()` é chamada, então, por padrão, o programa imprimirá na tela o texto `"Olá, visitante! Seja bem-vindo(a)!"`.
- Observação: não é necessário que a função esteja no começo do nosso programa. Podemos movê-la para baixo, mas precisamos dizer ao interpretador que temos uma função principal e uma função `saudacao()`.

````python
def main():
    nome = input("Qual é o seu nome? ").strip().title()
    saudacao(nome)

def saudacao(nome = 'visitante'):
    print(f"Olá, {nome}! Seja bem-vindo(a)!")

main()
````

### Parâmetros posicionais e nomeados

- Considere o exemplo a seguir:

````python
def apresentar(nome, curso):
    print(f"Meu nome é {nome} e o meu curso é {curso}")

apresentar("Arthur", "Economia")
````

O que acontece se chamarmos a função assim: `apresentar("Economia", "Arthur")`?  

- Nesse caso, a saída será **incorreta do ponto de vista semântico**:

    ```python
    Meu nome é Economia e o meu curso é Arthur
    ```

- Isso ocorre porque, em `apresentar(nome, curso)`, usamos **argumentos posicionais**: o Python associa os valores pela **ordem** em que são passados:
  
    1. primeiro argumento → `nome`
    2. segundo argumento → `curso`

- Portanto, nos argumentos posicionais, **a posição importa mais do que o significado do valor**.

- Para evitar esse tipo de erro, podemos usar **argumentos nomeados** (ou por palavra-chave), deixando explícito qual valor pertence a cada parâmetro:

````python
apresentar(nome="Arthur", curso="Economia")
apresentar(curso="Economia", nome="Arthur")
````

- Com argumentos nomeados, a correspondência é feita pelo **nome do parâmetro**, não pela posição, tornando o código mais claro e menos sujeito a trocas acidentais.

### 📊 Exemplo Econômico: Juros compostos

A fórmula de juros compostos é um conceito central em finanças:

$$M = C \cdot (1 + i)^t$$

onde:
- **M** = Montante final
- **C** = Capital inicial (principal)
- **i** = Taxa de juros por período
- **t** = Número de períodos

O parâmetro padrão `periodos=1` é útil quando queremos calcular o rendimento de apenas um período sem precisar especificar esse argumento toda vez.

````python
def juros_compostos(capital, taxa, periodos=1):
    montante = capital * (1 + taxa) ** periodos
    rendimento = montante - capital
    print(f"Capital inicial:  R$ {capital:,.2f}")
    print(f"Taxa por período: {taxa*100:.1f}%")
    print(f"Períodos:         {periodos}")
    print(f"Rendimento:       R$ {rendimento:,.2f}")
    print(f"Montante final:   R$ {montante:,.2f}")


print("--- Rendimento da poupança em 1 mês ---")
juros_compostos(capital=10_000, taxa=0.005)  # usa o padrão periodos=1

print()
print("--- Rendimento do Tesouro Selic em 12 meses ---")
juros_compostos(capital=10_000, taxa=0.005, periodos=12)
````

## Retornando valores <a id="retorno"></a>

Até agora, nossas funções apenas exibiam resultados com `print()`, mas não retornavam nenhum valor. A palavra-chave `return` permite que uma função **retorne um resultado** para quem a chamou, possibilitando seu uso em expressões e cálculos.

### `print()` vs. `return`: qual a diferença?

````python
# Versão com print: o resultado é exibido, mas não pode ser reutilizado
def dobro_print(n):
    print(n * 2)

# Versão com return: o resultado é devolvido e pode ser usado em outros cálculos
def dobro_return(n):
    return n * 2

# Com print: funciona para exibir, mas não dá para usar o valor depois
resultado_print = dobro_print(5)   # exibe 10 na tela...
print(resultado_print)             # ...mas a variável guarda None!

print()
# Com return: o valor é capturado e pode ser usado em novos cálculos
resultado_return = dobro_return(5)
print(resultado_return)            # 10
print(resultado_return + 3)        # 13 — podemos continuar operando!
````

> 💡 **Regra prática:** use `print()` quando quiser apenas **mostrar** algo ao usuário. Use `return` quando quiser que a função **produza um valor** que será usado em outro lugar do programa. Para construir modelos econômicos com vários cálculos encadeados, `return` é indispensável.

No exemplo a seguir, o valor é passado como parâmetro à função, que realiza o cálculo e retorna o resultado para a função principal:

````python
def main():
    x = int(input("Qual o valor de x? "))
    print("x ao quadrado é:", quadrado(x))

def quadrado(n):
    return n * n

main()
````

### 📊 Exemplo Econômico: Encadeando cálculos com `return`

Suponha que queremos calcular o **salário líquido** de um trabalhador, descontando INSS e IRPF. Em vez de fazer tudo de uma vez, podemos criar funções separadas e encadeá-las:

- `calcular_inss(salario_bruto)` → retorna o desconto do INSS
- `calcular_irpf(base_irpf)` → retorna o desconto do IRPF
- `salario_liquido(salario_bruto)` → usa as duas funções anteriores para chegar ao valor final

Esse é um exemplo de **modularização**: cada função faz uma coisa só, e elas se combinam para resolver um problema maior.

````python
# Alíquota simplificada para fins didáticos
def calcular_inss(salario_bruto):
    teto_inss = 8_475.55  # teto do INSS em 2026
    salario_inss = min(salario_bruto, teto_inss)

    faixas_inss = [
        (1_621.00, 0.075),
        (2_902.84, 0.09),
        (4_354.27, 0.12),
        (teto_inss, 0.14)
    ]

    inss = 0
    anterior = 0
    for limite, aliquota in faixas_inss:
        if salario_inss > limite:
            inss += (limite - anterior) * aliquota
            anterior = limite
        else:
            inss += (salario_inss - anterior) * aliquota
            break
    return inss

def calcular_irpf(base_irpf):
    # Tabela mensal do IRPF vigente a partir de maio de 2025
    faixas_irpf = [
        (2_428.80, 0.00, 0.00),
        (2_826.65, 0.075, 182.16),
        (3_751.05, 0.15, 394.16),
        (4_664.68, 0.225, 675.49),
        (float("inf"), 0.275, 908.73),
    ]

    irpf = 0
    for limite, aliquota, deducao in faixas_irpf:
        if base_irpf <= limite:
            irpf = (base_irpf * aliquota) - deducao
            break

    return max(irpf, 0)  # IR não pode ser negativo

def salario_liquido(salario_bruto):
    desconto_inss = calcular_inss(salario_bruto)
    base_irpf = salario_bruto - desconto_inss
    desconto_irpf = calcular_irpf(base_irpf)
    liquido = salario_bruto - desconto_inss - desconto_irpf
    return liquido

# Testando para diferentes faixas salariais
salarios = [1_500, 3_000, 6_000, 12_000]

print(f"{'Salário Bruto':>16} | {'INSS':>10} | {'IRPF':>10} | {'Salário Líquido':>16}")
print("-" * 62)

for bruto in salarios:
    inss = calcular_inss(bruto)
    irpf = calcular_irpf(bruto - inss)
    liquido = salario_liquido(bruto)
    print(f"R$ {bruto:>12,.2f} | R$ {inss:>7,.2f} | R$ {irpf:>7,.2f} | R$ {liquido:>13,.2f}")
````

### Funções com múltiplos retornos

- Funções podem retornar **mais de um valor** usando uma única instrução `return`, separando os valores por vírgula.
- Na prática, o Python agrupa esses valores em uma **tupla**.
- Ao chamar a função, podemos fazer o **desempacotamento** em variáveis diferentes.
- Isso é útil quando uma mesma função produz mais de um resultado relacionado (por exemplo, soma e produto).

````python
def operacoes(a, b):
    soma = a + b
    produto = a * b
    return soma, produto


s, p = operacoes(2, 3)
print(s, p)  # Saída: 5 6
````

- Também é possível retornar outros tipos de objetos, como listas e dicionários:

````python
import statistics

def descritivas(x):
    media = statistics.mean(x)
    desvio_padrao = statistics.stdev(x)
    return {'media': media, 'desvio_padrao': desvio_padrao}

dados = [1, 2, 3, 4, 5]
resultados = descritivas(dados)
print(resultados)
````

### 📊 Exemplo Econômico: Métricas de inflação

Na análise econômica, frequentemente precisamos calcular várias estatísticas ao mesmo tempo. O exemplo abaixo calcula a **inflação acumulada** e a **inflação média mensal** a partir de uma lista de taxas mensais, retornando ambas de uma única função.

````python
def analisar_inflacao(taxas_mensais):
    """
    Recebe uma lista de taxas mensais de inflação (em decimal)
    e retorna a inflação acumulada e a média mensal.
    """
    # Inflação acumulada: produto de (1 + taxa) para cada mês
    acumulada = 1
    for taxa in taxas_mensais:
        acumulada *= (1 + taxa)
    acumulada = acumulada - 1  # Converte de volta para percentual

    # Inflação média mensal
    media_mensal = statistics.mean(taxas_mensais)

    return acumulada, media_mensal


# IPCA mensal hipotético (em decimal)
ipca_2024 = [0.0042, 0.0083, 0.0016, 0.0038, 0.0044,
             0.0050, 0.0038, 0.0044, 0.0044, 0.0056,
             0.0039, 0.0052]

acum, media = analisar_inflacao(ipca_2024)
print(f"Inflação acumulada no ano: {acum*100:.2f}%")
print(f"Inflação média mensal:     {media*100:.2f}%")
````

## Aplicação: Teorema Central do Limite

Na aula anterior, escrevemos um código para simular o **Teorema Central do Limite (TCL)**: ao retirar muitas amostras de uma distribuição e calcular a média de cada uma, a distribuição dessas médias tende a ser normal — independentemente da distribuição original.

O código original funcionava, mas tinha uma limitação: os parâmetros da simulação (número de amostras, tamanho da amostra, distribuição) estavam "fixados" diretamente no código. Para testar cenários diferentes, era preciso reescrever o código toda vez.

Veja a diferença ao transformá-lo em uma **função reutilizável**:

````python
# Código original (da aula anterior) — parâmetros fixos, não reutilizável
import numpy as np

np.random.seed(1234)
n_amostras = 10000
tamanho_amostra = 5

medias = []
for _ in range(n_amostras):
    amostra = np.random.uniform(0, 1, tamanho_amostra)
    medias.append(np.mean(amostra))

media_das_medias = np.mean(medias)
print(f"Média das médias: {media_das_medias:.4f}")
````

Agora, transformamos o código em uma função. Isso nos permite:

- Testar facilmente diferentes tamanhos de amostra e números de repetições
- Reutilizar a lógica em outras partes do programa
- Documentar claramente o que cada parâmetro faz

Repare também como os **parâmetros com valores padrão** (`n_amostras=10000`, `tamanho_amostra=30`, `seed=1234`) deixam a função flexível:

- Ela funciona sem nenhum argumento, mas também aceita configurações personalizadas.

````python
import numpy as np

def simular_tcl(n_amostras=10000, tamanho_amostra=30, seed=1234):
    np.random.seed(seed)
    medias = []

    for _ in range(n_amostras):
        amostra = np.random.uniform(0, 1, tamanho_amostra)
        medias.append(np.mean(amostra))

    return np.mean(medias)

# Chamada simples com valores padrão
resultado = simular_tcl()
print(f"Média das médias (padrão): {resultado:.4f}")
````

### Verificando a convergência do TCL

Uma das demonstrações do código anterior era observar como a média das médias converge para o valor teórico (0.5 no caso da Uniforme[0,1]) conforme aumentamos o número de amostras. Com a função, isso fica muito mais limpo e legível:

````python
# Comparando a convergência para diferentes números de amostras
# (equivalente ao segundo bloco da aula anterior, agora reutilizando a função)
lista_n_amostras = [1, 10, 100, 1_000, 10_000, 100_000]

print(f"{'Nº de amostras':>16} | {'Média das médias':>18} | {'Erro vs. 0.5':>14}")
print("-" * 56)

for n in lista_n_amostras:
    media = simular_tcl(n_amostras=n, tamanho_amostra=5)
    erro = abs(media - 0.5)
    print(f"{n:>16,} | {media:>18.4f} | {erro:>14.4f}")
````

Observe como, ao aumentar o número de amostras, a média das médias se aproxima cada vez mais do valor teórico esperado (0.5). Esse é exatamente o TCL em ação — e com a função, podemos explorar esse comportamento variando os parâmetros com apenas uma linha de código.
