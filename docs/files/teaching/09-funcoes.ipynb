Até aqui, nossos programas seguiam um fluxo linear: o Python executava cada instrução na ordem em que ela aparecia. Essa abordagem funciona bem para tarefas simples, mas rapidamente se torna problemática à medida que os programas crescem.

Imagine que você precisa calcular o PIB pela ótica da despesa em vinte cenários diferentes. Sem funções, você repetiria a mesma fórmula vinte vezes — e bastaria um erro em uma das cópias para comprometer toda a análise.

**Funções** resolvem esse problema: elas permitem que um bloco de código seja escrito uma única vez e reutilizado quantas vezes for necessário, com diferentes valores de entrada.

## O que são funções? <a id="funcoes"></a>

- Funções são blocos de código reutilizáveis que executam uma tarefa específica.
- Permitem organizar, reutilizar e dividir o código em partes menores e mais compreensíveis.

### Por que usar funções?

- Evitam repetição de código.
- Facilitam a manutenção e o entendimento do programa.
- Permitem reutilização e modularização.

### Sintaxe básica de uma função

- Para definir uma nova função no Python, usamos a palavra-chave `def`, seguida do nome da função e, opcionalmente, pelos parâmetros.
- Entre parênteses, definimos os **parâmetros** que a função recebe.
- Todo o bloco de código **indentado** abaixo do cabeçalho da função fará parte dela.
- Embora não seja obrigatório, as funções podem retornar algum resultado.

A estrutura básica de uma função é a seguinte:

- `def`: Palavra-chave que indica a definição de uma função.
- `nome_da_funcao`: Nome descritivo que indica o que a função faz.
- `parametros`: Valores que a função recebe para processar (opcional).
- `return`: Palavra-chave que indica o valor que a função devolve após sua execução (opcional).

```python
def nome_da_funcao(parametros):
    <instrucoes>
    ...
    return <valor> (opcional)
```

## Parâmetros de uma função <a id="parametros"></a>

### Parâmetros posicionais e nomeados

- Considere o exemplo a seguir:

O que acontece se chamarmos a função assim: `apresentar("Economia", "Arthur")`?  

- Nesse caso, a saída será **incorreta do ponto de vista semântico**:

    ```python
    Meu nome é Economia e o meu curso é Arthur
    ```

- Isso ocorre porque, em `apresentar(nome, curso)`, usamos **argumentos posicionais**: o Python associa os valores pela **ordem** em que são passados:
  
    1. primeiro argumento → `nome`
    2. segundo argumento → `curso`

- Portanto, nos argumentos posicionais, **a posição importa mais do que o significado do valor**.

- Para evitar esse tipo de erro, podemos usar **argumentos nomeados** (ou por palavra-chave), deixando explícito qual valor pertence a cada parâmetro:

- Com argumentos nomeados, a correspondência é feita pelo **nome do parâmetro**, não pela posição, tornando o código mais claro e menos sujeito a trocas acidentais.

## Retornando valores <a id="retorno"></a>

Até agora, nossas funções apenas exibiam resultados com `print()`, mas não retornavam nenhum valor. A palavra-chave `return` permite que uma função **retorne um resultado** para quem a chamou, possibilitando seu uso em expressões e cálculos.

### `print()` vs. `return`: qual a diferença?

> 💡 **Regra prática:** use `print()` quando quiser apenas **mostrar** algo ao usuário. Use `return` quando quiser que a função **produza um valor** que será usado em outro lugar do programa. Para construir modelos econômicos com vários cálculos encadeados, `return` é indispensável.

### 📊 Exemplo Econômico: Encadeando cálculos com `return`

Suponha que queremos calcular o **salário líquido** de um trabalhador, descontando INSS e IRPF. Em vez de fazer tudo de uma vez, podemos criar funções separadas e encadeá-las:

- `calcular_inss(salario_bruto)` → retorna o desconto do INSS
- `calcular_irpf(base_irpf)` → retorna o desconto do IRPF
- `salario_liquido(salario_bruto)` → usa as duas funções anteriores para chegar ao valor final

Esse é um exemplo de **modularização**: cada função faz uma coisa só, e elas se combinam para resolver um problema maior.

### Funções com múltiplos retornos

- Funções podem retornar **mais de um valor** usando uma única instrução `return`, separando os valores por vírgula.
- Na prática, o Python agrupa esses valores em uma **tupla**.
- Ao chamar a função, podemos fazer o **desempacotamento** em variáveis diferentes.
- Isso é útil quando uma mesma função produz mais de um resultado relacionado (por exemplo, soma e produto).

## Aplicação: Teorema Central do Limite

Na aula anterior, escrevemos um código para simular o **Teorema Central do Limite (TCL)**: ao retirar muitas amostras de uma distribuição e calcular a média de cada uma, a distribuição dessas médias tende a ser normal — independentemente da distribuição original.

O código original funcionava, mas tinha uma limitação: os parâmetros da simulação (número de amostras, tamanho da amostra, distribuição) estavam "fixados" diretamente no código. Para testar cenários diferentes, era preciso reescrever o código toda vez.

Veja a diferença ao transformá-lo em uma **função reutilizável**:

````python
# Código original (da aula anterior) — parâmetros fixos, não reutilizável
import numpy as np

np.random.seed(1234)
n_amostras = 10000
tamanho_amostra = 30

medias = []
for _ in range(n_amostras):
    amostra = np.random.uniform(0, 1, tamanho_amostra)
    medias.append(np.mean(amostra))

media_das_medias = np.mean(medias)
print(f"Média das médias: {media_das_medias:.4f}")
````

Agora, transformamos o código em uma função. Isso nos permite:
- Testar facilmente diferentes tamanhos de amostra e números de repetições
- Reutilizar a lógica em outras partes do programa
- Documentar claramente o que cada parâmetro faz

Repare também como os **parâmetros com valores padrão** (`n_amostras=10000`, `tamanho_amostra=30`, `seed=19`) deixam a função flexível: ela funciona sem nenhum argumento, mas também aceita configurações personalizadas.

### Verificando a convergência do TCL

Uma das demonstrações do código anterior era observar como a média das médias converge para o valor teórico (0.5 no caso da Uniforme[0,1]) conforme aumentamos o número de amostras. Com a função, isso fica muito mais limpo e legível:

Observe como, ao aumentar o número de amostras, a média das médias se aproxima cada vez mais do valor teórico esperado (0.5). Esse é exatamente o TCL em ação — e com a função, podemos explorar esse comportamento variando os parâmetros com apenas uma linha de código.
