## O que é o Pandas? <a id="pandas"></a>

**Pandas** é uma biblioteca Python criada especificamente para manipulação e análise de dados tabulares. O nome vem de *panel data* — termo muito familiar para economistas.

Pense no Pandas como uma planilha Excel turbinada: você pode filtrar linhas, criar novas colunas, calcular médias por grupo, combinar bases de dados — tudo com poucas linhas de código e de forma reproduzível.

### Por que Pandas em Economia?

Em quase todo projeto empírico precisamos:
- Importar bases de dados;
- Selecionar variáveis relevantes e períodos de interesse;
- Tratar valores ausentes e inconsistências;
- Calcular indicadores derivados (valores per capita, taxas de variação, etc.);
- Combinar bases distintas;
- Resumir dados por grupos (médias estaduais, medianas por setor, etc.)

O Pandas resolve tudo isso de forma eficiente e legível.

### Operações com Series

Diferente dos arrays do NumPy, operações entre Series (`+`, `-`, `/`, `*`) alinham os valores pelos seus **rótulos** (e não pela posição). 

O índice do resultado será a união ordenada dos índices das duas Series. Isso permite realizar operações mesmo quando os rótulos não coincidem totalmente, tornando o trabalho com dados rotulados muito mais flexível.

```python
s1 = pd.Series(
    data = range(4),
    index = ["A", "B", "C", "D"]
)

s2 = pd.Series(
    data = range(10, 14),
    index = ["B", "C", "D", "E"]
)

...
```

Como você pode ver acima, os índices que coincidem são operados normalmente. Já os índices que não existem em ambas as Series aparecem no resultado com valor `NaN` (ausente).

Isso é útil para combinar informações de fontes diferentes sem perder o alinhamento dos dados.

Apesar de flexível, recomenda-se evitar o uso do tipo de dado "object" em Series, pois ele consome mais memória. Isso acontece porque, em uma Series do tipo "object", cada elemento armazena informações sobre seu próprio tipo individualmente.

Podemos inspecionar o tipo de cada elemento em uma Series mista usando o método `map`:

Note que cada objeto na nossa Series pode ter um tipo diferente. Essa flexibilidade tem um custo. Compare o uso de memória das Series abaixo:

```python
x1 = pd.Series([1, 2, 3])
print(f"x1 dtype: {x1.dtype}")
print(f"x1 Uso da memória: {x1.memory_usage(deep=True)} bytes")
print()
x2 = pd.Series([1, 2, "3"])
print(f"x2 dtype: {x2.dtype}")
print(f"x2 Uso da memória: {x2.memory_usage(deep=True)} bytes")
```

**Resumindo**

Sempre que possível, utilize tipos de dados uniformes em suas Series — eles são muito mais eficientes em termos de memória!

Outro ponto importante: o valor `nan` (usado para representar dados ausentes) é do tipo float. Isso pode causar problemas se você tiver uma Series de inteiros e um valor ausente, pois o Pandas irá converter toda a Series para float automaticamente.

Recentemente, o Pandas implementou um tipo especial chamado "nullable integer dtype", que permite lidar com valores ausentes em Series de inteiros sem precisar converter tudo para float. Note o "I" maiúsculo em `Int64`, que diferencia esse tipo do `int64` do NumPy:

```python
# Tipo especial para valores faltantes
type(np.nan)
print(pd.Series([1, 2, 3, np.nan]))
print(pd.Series([1, 2, 3, np.nan]).astype('Int64'))
```

## Importando Dados <a id="importacao"></a>

Na prática, raramente criamos DataFrames manualmente — quase sempre importamos arquivos externos. O Pandas suporta diferentes formatos de dados, tornando o processo de leitura muito flexível e eficiente.

A função mais utilizada é `pd.read_csv()`, que lê arquivos CSV ou TXT. Veja outros formatos suportados:

| Função               | Formato                |
|----------------------|------------------------|
| `pd.read_csv()`      | CSV / TXT              |
| `pd.read_excel()`    | Excel (.xlsx, .xls)    |
| `pd.read_json()`     | JSON                   |
| `pd.read_parquet()`  | Parquet (big data)     |
| `pd.read_html()`     | Tabelas HTML           |

A função `pd.read_csv()` possui diversos argumentos para facilitar a importação de dados de acordo com a sua necessidade. Os principais são:

| Argumento    | O que faz                    | Exemplo                      |
|--------------|------------------------------|------------------------------|
| `sep`        | Define o separador de colunas| `sep=';'`                    |
| `encoding`   | Codificação de caracteres    | `encoding='utf8'`            |
| `decimal`    | Separador decimal            | `decimal=','`                |
| `index_col`  | Coluna a usar como índice    | `index_col='ano'`            |
| `usecols`    | Importar só certas colunas   | `usecols=['pais','ano']`     |
| `nrows`      | Número de linhas a ler       | `nrows=1000`                 |

Para explorar todos os argumentos disponíveis, utilize `help(pd.read_csv)`.
