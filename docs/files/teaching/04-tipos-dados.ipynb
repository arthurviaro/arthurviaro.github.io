## Introdução às Funções <a id="funcoes"></a>

- Em praticamente toda linguagem de programação, temos acesso a funções.
- Uma função pode ser entendida como uma **ação** (um "verbo") que executa alguma tarefa no programa e retorna algum resultado.
- Toda linguagem já possui um conjunto de funções pré-definidas — ações básicas que o computador já sabe realizar. O programador pode utilizá-las para instruir o computador a executar determinadas operações.
- No nosso primeiro programa, utilizamos a função `print`. Essa função exibe algo na tela. Ela não imprime um texto fixo: ela imprime **aquilo que você passar como argumento**.

### Argumentos <a id="argumentos"></a>

- Um argumento é uma informação passada para uma função que influencia seu comportamento.
- Por exemplo:

```python
print("Olá, Mundo!")
```

- Aqui, `"Olá, Mundo!"` é o argumento da função `print`.
- Os criadores do Python não sabiam o que você desejaria exibir na tela. Por isso, projetaram a função `print` para receber um texto (`string`) como entrada, permitindo que você escolha o que será exibido.
- Nosso programa pode ser melhorado. Em vez de mostrar apenas `"Olá, Mundo!"`, podemos torná-lo interativo. Por exemplo, queremos que ele pergunte o nome do usuário e exiba na tela.
- Para isso, utilizamos a função `input`:

```python
input("Qual é o seu nome? ")
print("Olá, Mundo!")
```

- A função `input` exibe uma mensagem (prompt) e espera que o usuário digite algo.
- No entanto, esse código ainda não usa o nome digitado pelo usuário. Para isso, precisamos introduzir **variáveis**.

## Variáveis <a id="variaveis"></a>

- Uma variável é um espaço na memória que armazena um valor.
- Podemos modificar o programa da seguinte forma:

```python
nome = input("Qual é o seu nome? ")
print("Olá, Mundo!")
```

- O símbolo `=` em Python não significa igualdade matemática, mas sim **atribuição**.
- Ou seja, `name = input("Qual o valor de  your name? ")` significa: "Pegue o valor retornado por `input` e armazene na variável `nome`."
- Note que se fizermos:

```python
nome = input("Qual é o seu nome? ")
print("Olá, nome!")
```

- O programa exibirá literalmente: `Olá, nome!` independentemente do texto que o usuário digitou.
- Isso ocorre porque "nome" está entre aspas, sendo tratado como texto.
- A forma correta de escrever o código é:

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")

# Exiba Olá e o nome digitado
print("Olá, " + nome)
```

- Algumas funções possuem mais de um argumento que influencia seu comportamento.
- Podemos usar a vírgulas `,` para passar múltiplos argumentos e editar o nosso código da seguinte maneira:

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")

# Exiba Olá e o nome digitado
print("Olá,", nome)
```

## Strings e Parâmetros <a id="strings"></a>

- Uma string (`str`) é uma sequência de caracteres (texto).
- O programa a seguir imprime o texto em duas linhas diferentes:

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")
print("Olá, ")
print(nome)
```

- Isso acontece porque a função `print` possui o argumento `end='\n'`.
- O `\n` indica que a função `print` automaticamente cria uma quebra de linha ao ser executada.
- Note que podemos fornecer um argumento diferente de modo que a função não crie uma nova linha ao ser executada!
- Podemos modificar o programa da seguinte maneira:

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")
print("Olá,", end="")
print(nome)
```

- Ao fornecer `end=""`, estamos sobrescrevendo o valor padrão de `end`, de modo que ele nunca cria uma nova linha após esta primeira instrução `print`. Fornecendo o nome como "Fulano", a saída na janela do terminal será Olá, Fulano!.
- Parâmetros, portanto, são argumentos que podem ser aceitos por uma função.
- Você pode aprender mais na [documentação](https://docs.python.org/3/library/functions.html#print) do Python sobre o `print`.

### Um Pequeno Problema com Aspas <a id="um-pequeno-problema-com-aspas"></a>

- Observe como adicionar aspas como parte da sua `string` é um desafio. `print("Olá, "amigo"")` não funcionará e o interpretador retornará um erro.
- Geralmente, existem duas abordagens para corrigir isso. Primeiro, você poderia simplesmente mudar as aspas para aspas simples.
- Outra abordagem, mais comumente utilizada, seria escrever `print("Olá, \"amigo\"")`.
- As barras invertidas dizem ao interpretador que o caractere seguinte deve ser tratado como uma aspa dentro da string, evitando um erro do interpretador.
- Provavelmente, a maneira mais elegante de usar strings seria a seguinte: 

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")
print(f"Olá, {nome}")
```

- Observe o `f` em `print(f"Olá, {nome}")`. Esse `f` é um indicador especial para o Python tratar essa string de uma maneira específica, diferente das abordagens anteriores que ilustramos nesta aula.

### Fatiamento de Strings (Slicing) <a id="slicing"></a>

- O fatiamento de string (*slicing*) permite que você "corte" uma string para obter uma subseção dela.
- A sintaxe básica utiliza colchetes e índices: `string[início:fim]`.
  - Índice de Início: Onde o fatiamento começa (incluído).
  - Índice de Fim: Onde o fatiamento termina (excluído).
- Assim como outras operações de sequência no Python, o fatiamento (*slicing*) utiliza indexação baseada em zero, onde o primeiro elemento está no índice 0.
- Pense no índice não como o número do caractere, mas como a distância (ou deslocamento) do início da string.
  - No índice 0, você percorreu 0 caracteres (está no início).
  - No índice 1, você saltou 1 caractere para chegar ao próximo.
- **Dica**: Se uma string tem comprimento n, o último índice será sempre n−1.
- Imagine a string `nome = "Python"`. Os índices são:

| Letra    | P  | y  | t  | h  | o  | n  |
| -------- | -- | -- | -- | -- | -- | -- |
| Índice   | 0  | 1  | 2  | 3  | 4  | 5  |
| Negativo | -6 | -5 | -4 | -3 | -2 | -1 |

```python
nome = "Python"

# Pega do índice 0 até o 2 (o 2 não entra)
print(nome[0:2])  # Saída: Py

# Se omitir o início, ele começa do zero
print(nome[:4])   # Saída: Pyth

# Se omitir o fim, ele vai até o último caractere
print(nome[2:])   # Saída: thon

# Pega os três últimos caracteres
print(nome[-3:])     # Saída: hon

# Pega todos caracteres exceto os dois últimos
print(nome[:-2])     # Saída: Pyth
```

- Você também pode adicionar um terceiro parâmetro: `string[início:fim:passo]`.
- O passo determina quantos caracteres o Python deve pular.
- O Python permite contar de trás para frente usando números negativos.
- O -1 é sempre o último caractere.

```python
telefone = "0123456789"

# Pega do 1 ao fim, pulando de 2 em 2
print(telefone[1:10:2])  # Saída: 13579

# Pega do início ao fim, pulando de 2 em 2
print(telefone[::2])  # Saída: 02468

# Um truque famoso: inverter uma string
print(nome[::-1])    # Saída: nohtyP
```

- O slicing é muito usado para limpar dados formatados.
- Por exemplo, se você recebe uma data no formato `05-03-2026` e quer extrair apenas o ano:

```python
data = "05-03-2026"
ano = data[-4:]
print(ano)  # Saída: 2026
```

### Métodos de Strings <a id="str-metodos"></a>

- Você nunca deve esperar que seu usuário coopere conforme o pretendido. Portanto, você precisará garantir que a entrada (input) do seu usuário seja corrigida ou verificada.
- Acontece que as `strings` já possuem, de forma nativa, a capacidade de remover espaços em branco.
- Ao utilizar o método `strip` na variável nome (por exemplo, `nome = nome.strip()`), você removerá qualquer espaço em branco à esquerda e à direita da entrada do usuário.
- Um método é em essência uma sequência de instruções encapsuladas dentro de um único comando que recebe argumentos e devolve um valor.
- Você pode modificar seu código para:

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")

# Remove espaços em branco da string
nome = nome.strip()

# Print the output
print(f"Olá, {nome}")
```

- Ao executar este programa novamente, independentemente de quantos espaços você digitar antes ou depois do nome, ele removerá (`strip`) todo o espaço em branco.
- Usando o método `title`, o nome do usuário ficaria com as iniciais em maiúsculo (formato de título):

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")

# Remove espaços em branco da string
nome = nome.strip()

# Letras maiúsculas na primeira letra de cada palavra
nome = nome.title()

# Exibe o resultado
print(f"Olá, {nome}")
```

- Observe que você pode modificar seu código para ser mais eficiente:

```python
# Pergunte o nome do usuário
nome = input("Qual é o seu nome? ")

# Remove espaços em branco da string e coloca a primeira letra de cada palavra em maiúscula
nome = nome.strip().title()

# Exibe o resultado
print(f"Olá, {nome}")
```

- Nós poderíamos ir ainda mais longe!

```python
# Pergunte o nome do usuário, remova espaços em branco da string e coloque a primeira letra de cada palavra em maiúscula
nome = input("Qual é o seu nome? ").strip().title()

# Exibe o resultado
print(f"Olá, {nome}")
```

- Além do `strip()` e `title()`, você usará frequentemente outros métodos para normalizar dados.
- Por exemplo, converter todos os caracteres para maiúsculas ou minúsculas (`upper()` e `lower()`).
- Úteis para comparar textos sem se preocupar se o usuário digitou em maiúsculas ou minúsculas.

- `upper()`: Transforma toda a string em MAIÚSCULAS.
- `lower()`: Transforma toda a string em minúsculas.

```python
curso = "métodos computacionais em economia"

print(curso.lower())
print(curso.upper())
```

- Dividir uma string em partes menores (`split()`).
- Esse método "quebra" a string sempre que encontra o delimitador especificado e retorna uma lista contendo os trechos resultantes.
- No exemplo a seguir, o espaço `" "` é utilizado como critério de separação.

```python
curso = "métodos computacionais em economia"
print(curso.split(" "))
```

- Substituir todas as ocorrências de um trecho de texto por outro (`replace()`).

```python
texto = "Eu gosto de Java"

print(texto) # Saída: Eu gosto de Java
print(texto.replace("Java", "Python"))  # Saída: Eu gosto de Python
```

- Você pode aprender mais sobre strings e seus métodos na documentação oficial do Python sobre o tipo [str](https://docs.python.org/3/library/stdtypes.html#str)

## Inteiros (int) <a id="inteiros"></a>

- No Python, um número inteiro é conhecido como um `int`.
- No mundo da matemática, estamos familiarizados com os operadores `+, -, *, / e %`. 
- Esse último operador, `%` ou módulo, pode não ser muito familiar para você.
- Primeiro, podemos declarar algumas variáveis:

```python
x = 1
y = 2
z = x + y
print(z)
```

- Naturalmente, quando executamos o programa, obtemos o resultado 3 na janela do terminal. 
- Podemos tornar isso mais interativo usando a função `input`.

```python
x = input("Qual o valor de x? ")
y = input("Qual o valor de y? ")
z = x + y
print(z)
```

- Ao executar este programa, descobrimos que a saída está incorreta. Por que isso acontece?
- Anteriormente, vimos como o sinal de `+` concatena (junta) duas strings. Como a entrada do seu teclado chega ao interpretador como texto, ela é tratada como uma string.
- Portanto, precisamos converter essa entrada de uma string para um inteiro. Podemos fazer isso da seguinte forma:

```python
x = input("Qual o valor de x? ")
y = input("Qual o valor de y? ")
z = int(x) + int(y)
print(z)
```

- O resultado agora está correto. O uso de `int(x)` é chamado de "*casting*" (conversão de tipo), onde um valor é temporariamente alterado de um tipo de variável (neste caso, uma string) para outro (aqui, um inteiro).
- Podemos melhorar ainda mais nosso programa da seguinte forma:

```python
x = int(input("Qual o valor de x? "))
y = int(input("Qual o valor de y? "))
print(x + y)
```

- Isso ilustra que você pode executar funções dentro de funções. A função interna (*inner function*) é executada primeiro e, depois, a externa (*outer function*).
- Primeiro, a função `input` é executada. Em seguida, a função `int`.
- Você pode aprender mais na documentação do Python sobre o `int`.

## Floats (Pontos Flutuantes) <a id="float"></a>

- Um valor de ponto flutuante é um número real que possui um ponto decimal, como 0.52.
- Você pode alterar seu código para suportar floats da seguinte forma:

```python
x = float(input("Qual o valor de x? "))
y = float(input("Qual o valor de y? "))
print(x + y)
```

- Essa mudança permite que seu usuário digite 1.2 e 3.4 para apresentar um total de 4.6.
- Imaginemos, entretanto, que você queira arredondar o total para o número inteiro mais próximo. Olhando a documentação do Python para `round`, você verá que os argumentos disponíveis são `round(number[, ndigits])`.
- Esses colchetes indicam que algo **opcional** pode ser especificado pelo programador.
- Portanto, você poderia fazer `round(n)` para arredondar um dígito para o inteiro mais próximo.
- Alternativamente, você poderia codificar da seguinte forma:

```python
# Obtém a entrada do usuário
x = float(input("Qual o valor de x? "))
y = float(input("Qual o valor de y? "))

# Cria um resultado arredondado
z = round(x + y)

# Imprime o resultado
print(z)
```

- A saída será arredondada para o número inteiro mais próximo.
- E se quiséssemos formatar a saída de números longos? Por exemplo, em vez de ver 1000, você pode desejar ver 1,000.
- Você poderia modificar seu código assim:

```python
# Obtém a entrada do usuário
x = float(input("Qual o valor de x? "))
y = float(input("Qual o valor de y? "))

# Cria um resultado arredondado
z = round(x + y)

# Imprime o resultado formatado
print(f"{z:,}")
```

- Embora pareça enigmático, esse `print(f"{z:,}")` cria um cenário onde o z exibido incluirá vírgulas (separador de milhar), resultando em algo como 1,000 ou 2,500.

### Mais sobre floats <a id="mais-sobre-floats"></a>

- Como podemos arredondar valores de ponto flutuante?
- Primeiro, modifique seu código:

```python
# Obtém a entrada do usuário
x = float(input("Qual o valor de x? "))
y = float(input("Qual o valor de y? "))

# Calcula o resultado
z = x / y

# Imprime o resultado
print(z)
```

- Ao inserir 2 como x e 3 como y, o resultado z é 0.6666666666, parecendo ir ao infinito como poderíamos esperar.
- Imaginemos que queremos arredondar isso. Poderíamos modificar nosso código:

```python
# Obtém a entrada do usuário e arredonda
x = float(input("Qual o valor de x? "))
y = float(input("Qual o valor de y? "))

# Calcula o resultado e arredonda para 2 casas
z = round(x / y, 2)

# Imprime o resultado
print(z)
```

- Como esperado, isso arredondará o resultado para as duas casas decimais mais próximas.
- Também poderíamos usar uma `f-string` para formatar a saída da seguinte forma:

```python
# Obtém a entrada do usuário
x = float(input("Qual o valor de x? "))
y = float(input("Qual o valor de y? "))

# Calcula o resultado
z = x / y

# Imprime o resultado formatado
print(f"{z:.2f}")
```

- Você pode aprender mais na documentação do Python sobre [float](https://docs.python.org/3/library/functions.html#float).

## Booleanos <a id="booleanos"></a>

- Em Python, um booleano representa um de dois valores: `True` (Verdadeiro) ou `False` (Falso).
- Diferente de strings, eles não levam aspas e devem sempre começar com letra maiúscula.
- `True` pode ser convertido automaticamente para 1; `False` para 0. 
- Isso pode ser muito conveniente, por exemplo, ao contar valores verdadeiros ou falsos nas colunas de uma tabela de dados.

```python
resultado = True + True + False + True
print(resultado)
```

## Explorando o Python <a id="explorando-python"></a>

- Além de consultar a documentação online, você pode usar funções nativas para descobrir o que cada tipo de dado é capaz de fazer.

### Função dir() <a id="funcao-dir"></a>

- A função `dir` funciona como um "mapa". Ela lista todos os métodos e atributos disponíveis para um objeto ou tipo de dado.
- Se você quiser saber tudo o que pode fazer com uma string, basta digitar: `dir(str)`.
- Isso retornará uma lista gigante contendo nomes como `strip`, `title`, `lower`, `upper`, entre outros. É a forma mais rápida de verificar: "Será que existe um método para isso?".

### Função help() <a id="funcao-help"></a>

- Enquanto o `dir` diz o que existe, o `help` explica o que cada coisa faz.
- Se você esqueceu como usar o `strip`, por exemplo, pode chamar a ajuda interativa: `help(str.strip)`.
- O interpretador exibirá uma breve explicação técnica sobre como o método funciona, quais argumentos ele aceita e o que ele retorna.
