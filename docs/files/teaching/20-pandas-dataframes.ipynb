### Alterando o índice

Os **rótulos do índice** de um DataFrame identificam cada linha. Podemos trabalhar com eles de várias formas:

- **Transformar uma coluna em índice:**  
  Use `.set_index('coluna')` para tornar uma coluna o novo índice do DataFrame.

- **Dar um nome ao índice:**  
  Modifique o atributo `df.index.name` para deixar o índice mais informativo (ex: `df.index.name = 'id'`).

- **Restaurar o índice padrão:**  
  Use `.reset_index()` para mover o índice atual para uma coluna e voltar ao índice padrão (números inteiros sequenciais).

- **Alterar manualmente os valores do índice:**  
  Você pode modificar `df.index` diretamente para definir novos rótulos (menos comum, mas útil em casos específicos).

> **Resumo:**  
> O índice serve para identificar as linhas. Na maioria dos casos, usamos `.set_index()` e `.reset_index()` para manipular o índice de forma prática e segura.

### Selecionando e Filtrando Dados <a id="selecao-filtros"></a>

Selecionar colunas e filtrar linhas são as operações mais frequentes em qualquer análise de dados.

**Seleção de colunas**

Para selecionar uma única coluna, use `df['nome']`, que retorna uma `Series`. Para múltiplas colunas, passe uma lista: `df[['col1', 'col2']]`, que retorna um `DataFrame`:

```python
# Uma coluna → retorna uma Series
df['sexo'].head()
```

```python
# Múltiplas colunas → retorna um DataFrame
df[['uf', 'sexo', 'idade', 'renda_trabalho']].head()
```

**Filtros booleanos e `.loc[]`**

Para filtrar linhas, criamos uma **máscara booleana** — uma `Series` de `True`/`False` — e a usamos como índice. Para selecionar linhas **e** colunas ao mesmo tempo, usamos `.loc[linhas, colunas]`.

Ao combinar múltiplas condições, use `&` (e lógico) e `|` (ou lógico), com **cada condição entre parênteses**:

```python
# Máscara: mulheres (sexo == 2) ocupadas (ocupado == 1) em SP (uf == 35)
mascara = (df['uf'] == 35) & (df['sexo'] == 2) & (df['ocupado'] == 1)

# .loc[linhas, colunas] — seleciona linhas pela máscara e apenas as colunas desejadas
df.loc[mascara, ['uf', 'sexo', 'idade', 'escolaridade', 'renda_trabalho']].head()
```

**Métodos úteis para filtros**

O Pandas oferece atalhos que tornam os filtros mais concisos e legíveis:

- `.isin(lista)`: retorna `True` se o valor estiver em uma lista
- `.between(min, max)`: equivalente a `(col >= min) & (col <= max)`, mas mais legível
- `.query("expressão")`: recebe a condição como string, com sintaxe próxima ao SQL

```python
# .isin(): estados do Sudeste: SP (35), RJ (33), MG (31), ES (32)
df[df['uf'].isin([31, 32, 33, 35])][['uf', 'sexo', 'idade', 'renda_trabalho']].head()
```

```python
# .between(): pessoas em idade ativa (25 a 59 anos)
df[df['idade'].between(25, 59)][['uf', 'sexo', 'idade', 'anos_estudo', 'renda_trabalho']].head()
```

```python
# .query(): condições como string, muito legível para filtros compostos
df.query("uf in [31, 32, 33, 35] and idade >= 25 and renda_trabalho > 3000")[['uf', 'idade', 'renda_trabalho']].head()
```

### Mapeando variáveis qualitativas

> **Dica:** Muitas vezes, variáveis qualitativas vêm codificadas como números em bases de dados. Para facilitar a análise e a visualização, é interessante transformar esses códigos em rótulos mais informativos.

Podemos "mapear" uma variável qualitativa usando o método `.map()` e passando um dicionário como argumento. Isso torna os dados mais legíveis e evita confusões na interpretação.

No exemplo abaixo, vamos mapear a variável `sexo`, em que o código `1` representa "homem" e o código `2` representa "mulher":

```python
df['sexo'] = df['sexo'].map({1:'homem', 2:'mulher'})
df[(df['sexo'] == 'homem') | (df['sexo'] == 'mulher')]['sexo']
```

```python
df[df['sexo'] == 'homem']
```

### Calculando Frequência

O método `value_counts()` do Pandas é uma ferramenta poderosa usada para contar as ocorrências de valores únicos em uma Series ou DataFrame. Ele é amplamente utilizado em análise de dados para entender rapidamente a distribuição de dados categóricos ou discretos.

```python
# Frequência absoluta
df['sexo'].value_counts()
```

```python
# Frequência relativa
df['sexo'].value_counts(normalize = True).round(3)
```

```python
# Frequência relativa para um estado específico
df[df['uf'] == 33]['sexo'].value_counts(normalize = True).round(3)
```

Para cruzar a distribuição de **duas variáveis** qualitativas, usamos `pd.crosstab()`, que gera uma tabela de contingência. O argumento `normalize` permite normalizar por linha (`'index'`), por coluna (`'columns'`) ou pelo total (`True`):

```python
# Tabela de contingência: sexo x condição na força de trabalho (1=ativa, 2=inativa)
pd.crosstab(df['sexo'], df['forca_trabalho'])
```

```python
# Proporção dentro de cada grupo de sexo (normalize por linha)
pd.crosstab(df['sexo'], df['forca_trabalho'], normalize='index').round(3)
```

### Adicionando e removendo colunas

Existem duas formas principais de adicionar ou remover colunas em um DataFrame:

- Usar colchetes `[]` para criar ou modificar colunas.
- Usar o método `.drop()` para remover colunas.

Vamos importar uma cópia nova do conjunto de dados de exemplo para praticar.

```python
df
```

Podemos adicionar uma nova coluna a um DataFrame simplesmente usando colchetes `[]` com o nome da nova coluna e os valores desejados:

```python
df['fonte'] = 'IBGE'
df
```

Além de colunas com valores literais, podemos criar variáveis **derivadas** a partir das colunas existentes: transformações aritméticas, condições booleanas e categorizações. Isso é essencial na preparação de dados para análise.

**Nova coluna por expressão aritmética**

Qualquer operação vetorizada pode ser atribuída diretamente a uma nova coluna:

```python
df['renda_anual']   = df['renda_trabalho'] * 12
df[['sexo', 'idade', 'renda_trabalho', 'renda_anual']].dropna().head()
```

**Variável booleana**

Uma condição lógica aplicada a uma coluna retorna diretamente uma `Series` de `True`/`False`, que pode ser salva como nova variável:

```python
# Recebe acima do salário mínimo de 2025 (R$ 1.518)?
df['acima_sm'] = df['renda_trabalho'] >= 1518
df[['sexo', 'idade', 'renda_trabalho', 'acima_sm']].dropna().head(8)
```

**Variável categórica com `np.where()`**

`np.where(condição, valor_se_true, valor_se_false)` é um `if-else` **vetorizado** — aplica-se a todas as linhas de uma vez, sem loops. Para mais de duas categorias, aninhamos chamadas de `np.where()`:

```python
# Duas categorias: pessoa está em idade ativa (15 a 64 anos)?
df['em_idade_ativa'] = np.where(df['idade'].between(15, 64), 'ativa', 'inativa')
df[['idade', 'em_idade_ativa']].head(8)
```

```python
# Três categorias aninhadas: faixa etária
df['faixa_etaria'] = np.where(
    df['idade'] < 25, 'jovem (< 25)',
    np.where(df['idade'] < 60, 'adulto (25–59)', 'idoso (>= 60)')
)

df['faixa_etaria'].value_counts(normalize = True).round(3)
```

```python
# Faixa de renda em salários mínimos (SM = R$ 1.518 em 2025)
SM = 1518
df['faixa_renda'] = np.where(
    df['renda_trabalho'].isna(),          'sem renda declarada',
    np.where(df['renda_trabalho'] < SM,   'até 1 SM',
    np.where(df['renda_trabalho'] < 3*SM, '1 a 3 SM',
                                          'acima de 3 SM'))
)

df['faixa_renda'].value_counts(normalize = True).round(3)
```

## Mais Operações com DataFrame <a id="mais-operacoes"></a>

### Aplicando Funções Personalizadas

Nem sempre as operações prontas do Pandas (como `sum`, `mean`, etc.) são suficientes. Muitas vezes precisamos:

- aplicar uma transformação específica do seu problema
- usar funções de outras bibliotecas (como NumPy)
- ou criar uma lógica própria para tratar os dados

O Pandas permite aplicar essas funções diretamente nos dados, de forma simples e flexível, usando métodos como:

- `df.apply()`: aplica uma função por coluna ou por linha em um DataFrame (a função deve ser capaz de receber/retornar um array)
- `series.apply()` / `series.map()`: mesma ideia, mas para séries do Pandas

```python
df = pd.read_csv('pnad2025t4.csv')

df.columns = ['ano', 'trimestre', 'uf', 'capital', 'upa', 'estrato', 'numero_domicilio', 'painel', 'peso', 'projecao_populacao', 'dominios_projecao', 'sexo', 'idade', 'cor', 'escolaridade', 'anos_estudo', 'forca_trabalho', 'ocupado', 'tipo_ocupacao', 'atividade', 'renda_trabalho' ]

df
```

Por exemplo, suponha que você queira usar uma função do NumPy em uma coluna do seu DataFrame:

```python
df[['idade', 'renda_trabalho']].apply(np.log)
```

Ou você pode querer aplicar sua própria função personalizada:

```python
def log_renda(x):
    return np.log(1 + x)

df[['renda_trabalho']].apply(log_renda)
```

Também podemos usar a função lambda:

```python
df[['renda_trabalho']].apply(lambda x: np.log(1 + x))
```

Você pode até usar funções que exigem argumentos adicionais. Basta especificar os argumentos em `.apply()`:

```python
def log_renda(x, fator):
    return np.log(fator + x)

df[['renda_trabalho']].apply(log_renda, fator=1)
```

### Agrupando dados

O `.groupby()` é uma das ferramentas mais poderosas do Pandas para **análise exploratória**. Ele permite dividir o DataFrame em grupos com base nos valores de uma ou mais colunas, e aplicar funções de agregação (média, soma, contagem, etc.) separadamente em cada grupo.

A lógica segue o padrão **dividir → aplicar → combinar**:
1. **Dividir**: o DataFrame é particionado em subconjuntos por grupo
2. **Aplicar**: uma função é executada em cada subconjunto
3. **Combinar**: os resultados são reunidos em um novo DataFrame

Vamos agrupar os dados por UF (código da unidade da federação):

```python
dfg = df.groupby(by = 'uf')
dfg
```

O que é um objeto `DataFrameGroupBy`? 

Ele contém informações sobre os grupos do DataFrame. Na prática, um objeto retornado por `groupby` funciona como um dicionário que mapeia cada grupo aos índices das linhas correspondentes no DataFrame original. Ou seja, ele armazena quais linhas pertencem a cada grupo, permitindo que você acesse, analise e agregue dados facilmente por categoria.

Se quiser, você pode visualizar esse mapeamento de grupos acessando o atributo `.groups`:

```python
dfg.groups
```

```python
dfg.groups.keys
```

```python
# Conta linhas
dfg.size()
```

```python
# Conta valores não nulos por coluna
dfg.count()
```

Também podemos acessar diretamente os dados de um grupo específico usando o método `.get_group()`:

```python
dfg.get_group(35)
```

O uso mais comum, porém, é aplicar funções de agregação ao objeto retornado pelo `groupby`:

```python
dfg.mean().round(2)
```

```python
dfg[['anos_estudo', 'renda_trabalho']].mean().round(2)
```

Também podemos aplicar múltiplas funções usando o método `.aggregate()`:

```python
dfg.aggregate(['mean', 'std', 'count'])
```

Ou o método `.agg()`, o qual permite dar nomes personalizados às estatísticas:

```python
df.groupby(by = 'uf')['renda_trabalho'].agg(
    media    = 'mean',
    mediana  = 'median',
    desvpad  = 'std',
    maximo   = 'max',
    minimo   = 'min'
).round(2)
```

E ainda é possível aplicar funções diferentes para colunas diferentes:

```python
def num_range(x):
    return x.max() - x.min()

dfg.agg({
    "anos_estudo": ['max', 'min', 'mean', num_range],
    "renda_trabalho": ['mean']
}).round(2)
```

### Transformando os dados

Enquanto `.agg()` **reduz** o DataFrame (uma linha por grupo), o `.transform()` **preserva** o tamanho original — retorna um valor para cada linha, baseado no grupo dela.

**Caso de uso típico:** criar uma coluna com a média do grupo para comparar cada observação com sua referência no grupo.

```python
df['renda_media'] = dfg['renda_trabalho'].transform('mean')
df[['renda_trabalho', 'renda_media']]
```

### Tratando valores ausentes

Podemos combinar `.isnull()` com `.sum()` para contar os valores ausentes em cada coluna e com `.mean()` para obter a proporção:

```python
df.info()
```

Podemos remover valores ausentes com `.dropna()`:

```python
df.dropna()
```

Podemos imputar ("preencher") esses valores usando o método `.fillna()`. Esse método possui várias opções para preenchimento: você pode usar um valor fixo, a média da coluna, o valor anterior não nulo, etc.

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                  columns=list('ABCD'))
df
```

```python
# Preenche com 0
df.fillna(0)
```

```python
# Preenche com a média
df.fillna(df.mean())
```

```python
# Preenche com o próximo valor não nulo (backward fill)
df.bfill()
```

```python
# Preenche com o valor anterior não nulo (forward fill)
df.ffill()
```

```python
df.bfill().ffill()
```
